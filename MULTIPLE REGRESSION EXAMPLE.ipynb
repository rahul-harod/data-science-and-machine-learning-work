```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import statsmodels.api as sm
import seaborn as sns
sns.set()
```

```python
data1=pd.read_csv("F:\\1.03. Dummies.csv")
data1
```

```python
data=data1.copy()
#more than 75% attendance are yes=1
#convert categorical data to dummy data for account in regresstion
data['Attendance']=data['Attendance'].map({'Yes':1,'No':0})
data
```

```python
data.describe()
```

# REGRESSION

```python
y=data['GPA']
x1=data[['SAT','Attendance']]
```

```python
x=sm.add_constant(x1)
results=sm.OLS(y,x).fit()
results.summary()
```

```python
plt.scatter(data['SAT'],data['GPA'],c=data['Attendance'],cmap='RdYlGn_r')
yhat_no=0.6439+0.0014*data['SAT']
yhat_yes=0.6439+0.2226+0.0014*data['SAT']
yhat=0.275+data["SAT"]*0.0017
fig=plt.plot(data['SAT'],yhat_no,lw=2,c='#006837',label='regression line1')
fig=plt.plot(data['SAT'],yhat_yes,lw=2,c='#a50026',label='regression line2')
fig=plt.plot(data['SAT'],yhat,lw=4,c='orange',label='regression line')
plt.xlabel('SAT',fontsize=20)
plt.ylabel('GPA',fontsize=20)
plt.show
```
